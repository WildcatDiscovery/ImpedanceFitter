# Graph Dimensions
We can change the dimensions of the graph by calling on set_new_gph_dims(height, width) and customizing the size of the graph to our liking.


```python
#Say we want to make our graph 8 inches by 8 inches
ex_mpt.set_new_gph_dims(8,8)
ex_mpt.mpt_plot()
```

# Graph Window
We can not only change the size of the graph, but we can also customize the window of the graph we're looking at by calling on mpt plot, but this time, call out x_window = [x_min, x_max] and y_window = [y_min, y_max] and plant that into the plotting function.

```python
ex_mpt.mpt_plot(x_window = [0,5000], y_window = [0,5000])
```

# Fitting and Residuals
There are also two attributes in mpt_plot() that can be called after the fitting process has been fitted. The first is the fitting = 'on' attribute, which allows the user to view the Nyvquist Impedance graph with the fitting equation layered on top of it. It is off if it is not called. There is also the residual attribute, denoted by rr = 'on' which can also be switched on and off by typing the state into the mpt_plot() function. The residual attribute plots the original graph, and also plots another graph with the residuals of the fitted equation and the actual graph.

```python
#Turn on fitting option
#ONLY WORKS AFTER FITTING THE GRAPH
ex_mpt.mpt_plot(fitting = 'on')
```
